```python
import numpy as np
from matplotlib import pyplot as plt
plt.rcParams['xtick.labelsize']=20
plt.rcParams['ytick.labelsize']=20
```

```python
sigma = 5.0
r_max = 15.0
nr = 100
kp_max = 1.0

r = np.linspace(0.0,r_max,nr)
kp_r = np.exp(-r**2/sigma**2)
```

```python
phi_p_thresh = [0.2, 0.4, 0.6]
hill_exponent = np.arange(2,10)
phi_p_range = np.linspace(0.0,1.0)

for K in phi_p_thresh:
    fig,axs = plt.subplots(1,1,figsize=(10,6))
    for n in hill_exponent:
        rate_coeff = kp_max*phi_p_range**n/(phi_p_range**n + K**n)
        axs.plot(phi_p_range, rate_coeff,label='$n$ = ' + str(n))
    axs.set_title('$K_{threshold}$ = ' + str(K), fontsize=20)
    axs.legend()
    axs.set_xlabel(r'$\phi_p$', fontsize=20)
    axs.set_ylabel(r'$k_p \frac{\phi^n_p}{\phi^n_p + K^n}$',fontsize=20)
    plt.show()
```

```python
r = np.linspace(0.0,15.0,100)
f_r = np.exp(-r**2/5**2)
plt.plot(r,f_r)
plt.show()
```

```python
f_prime = np.gradient(f_r,r)
f_prime_prime = np.gradient(f_prime, r)
plt.plot(r, f_prime_prime)
plt.show()
```
