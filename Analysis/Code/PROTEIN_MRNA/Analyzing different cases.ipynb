```python
from matplotlib import pyplot as plt
import pandas as pd
plt.rcParams['xtick.labelsize']=20
plt.rcParams['ytick.labelsize']=20
```

```python
### Comparing the different equation forms
```

```python
directory = ["/home/npradeep96/Projects/lncRNAeffects/Output/ONLY_ACTIVITY/NEUMANN_BC/k_deg_0.01_phi0_0.15_larger_domain_flux_compk_p_0.01_reactiontype_2.0_thresh_0.0_FE_5.0_MP_1.0_Mmrna_0.05","/home/npradeep96/Projects/lncRNAeffects/Output/ONLY_ACTIVITY/NEUMANN_BC/k_deg_0.01_phi0_0.15_larger_domain_flux_compk_p_0.01_reactiontype_3.0_thresh_0.4_FE_5.0_MP_1.0_Mmrna_0.05"]
labels = ["Linear", "Hill"]
stats_file = "stats.txt"

fig, axs = plt.subplots(1,2,figsize=(20,8))

for i in range(len(directory)):
    file_name = directory[i] + '/' + stats_file
    df_stats = pd.read_csv(file_name, '\t')
    times = df_stats['t']
    steps = df_stats['step']
    phi_p = df_stats['phi_p_max']
    phi_m = df_stats['phi_m_avg']
    
    axs[0].plot(times, phi_p, '-o', label=labels[i])
    axs[0].set_ylabel('Protein concentration at enhancer well',fontsize=20)
    axs[0].set_xlabel('Time',fontsize=20)
    axs[1].plot(times, phi_m, '-o', label=labels[i],)
    axs[1].set_ylabel('Avg. mRNA concentration in system', fontsize=20)
    axs[1].set_xlabel('Time', fontsize=20)
    
    axs[0].legend(fontsize=20)
    axs[1].legend(fontsize=20)
    
plt.show()
    
```
